#### 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
df=pd.read_csv('Laliga.csv')
df.head()
```

```python
df.columns=df.iloc[0]
```

```python
df=df.drop(index=0)
df
```

```python
df= df.replace(to_replace='-', value = 0)
df
```

Inference- we replaced all the '-' by 0 using the replace function.

#### 2. Print all the teams which have started playing between 1930-1980.

```python
df['Debut'] = df['Debut'].str[:4]
```

```python
df.Debut.unique()
```

```python
df['Debut'] = df['Debut'].astype('int64')
```

```python
r1=df.loc[(df.Debut>=1930)&(df.Debut<=1980),['Team','Debut']]
r1
```

Inference- The above is the list of teams who started playing between 1930 and 1980.

#### 3. Print the list of teams which came Top 5 in terms of points

#### 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.
#### Goal_diff_count = GoalsFor - GoalsAgainst

```python
df['GoalsFor']=df['GoalsFor'].astype('int64')
df['GoalsAgainst']=df['GoalsAgainst'].astype('int64')
```

```python
def Goal_diff_count(Goals_For,Goals_Against):
    return Goals_For-Goals_Against
```

```python
df['Goal_diff']=df.apply(lambda x: Goal_diff_count(x['GoalsFor'],x['GoalsAgainst']),axis=1)
```

```python
df.head()
```

```python
df[df['Goal_diff']==df['Goal_diff'].max()]
```

Inference- Real Madrid is the team with maximum goal difference with 2807 goals.

```python
df[df['Goal_diff']==df['Goal_diff'].min()]
```

Inference- Racing Santander is the team with minimum goal difference with -525.

#### 5.  5. Create a new column with name “Winning Percent” and append it to the data set 

#### Percentage of Winning = (GamesWon / GamesPlayed)*100

#### If there are any numerical error, replace it with 0%


```python
df['GamesWon']=df['GamesWon'].astype('int64')
df['GamesPlayed']=df['GamesPlayed'].astype('int64')
```

```python
df['Winning Percent']=(df['GamesWon']/df['GamesPlayed'])*100
```

```python
df['Winning Percent']=df['Winning Percent'].fillna(value=0)
```

```python
df
```

Inferece- Added a new column 'Winning Percent'. And replaced the null values with 0.

#### Print the top 5 teams which has the highest Winning percentage

```python
r3=df.sort_values(["Winning Percent"],ascending=False)
```

```python
r3.loc[:,['Team','Winning Percent']].head()
```

Inference- List if the top 5 teams with highest winning percent.
